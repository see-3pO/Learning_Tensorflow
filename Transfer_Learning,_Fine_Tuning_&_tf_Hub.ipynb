```python
import tensorflow as tf
from tensorflow import keras
from keras import layers
import tensorflow_hub as hub
```

1. Pretrained Keras model

```python
# random data from a normal distribution
x = tf.random.normal(shape=(5, 299, 299, 3))
#  classes
y = tf.constant([0, 1, 2, 3, 4])
```

```python
model = keras.applications.InceptionV3(include_top=True)
model.summary()
```

```python
# accessing inputs and outputs of the desired layers
base_input = model.layers[0].input # input access
base_outputs = model.layers[-2].output # output access
# add a new dense layer
final_outputs = layers.Dense(5)(base_outputs)

# create the fine tuned model
new_model = keras.Model(inputs=base_input, outputs=final_outputs)
```

```python
new_model.compile(
    loss = keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer = keras.optimizers.Adam(),
    metrics = ['accuracy'],
)
```

```python
tf.config.run_functions_eagerly(True)
new_model.fit(x, y, epochs=10, verbose=2)
```

2. Tensorflow Hub

```python
# url from tf hub kaggle repo
inception_url = "https://kaggle.com/models/google/inception-v3/TensorFlow1/feature-vector/1"
```

```python
base_model = hub.KerasLayer(inception_url, input_shape=(299, 299, 3))
base_model.trainable = False

model = keras.Sequential([
    base_model,
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(5),
])
```

```python
model.compile(
    loss = keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer = keras.optimizers.Adam(),
    metrics = ['accuracy'],
)
```

```python
model.fit(x, y, epochs=10, verbose=2)
```
